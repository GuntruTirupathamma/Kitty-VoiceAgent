### Text to Speech conversion using GTTS

```python
!pip install gtts
```

```python
from gtts import gTTS
from IPython.display import Audio

text_to_speech = gTTS('''Hello everyone.
    Today is a new opportunity to learn and grow.
    Every small step we take builds confidence and experience.
    Let us stay curious, stay focused, and move forward with determination.''',lang='hi',tld='com')

text_to_speech.save('text_to_speech_gtts.wav')
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file, autoplay= False)
```

### Text to Speech conversion using pyttsx3

#### Changing the gender using pyttsx3

```python
!pip install pyttsx3
```

```python
import pyttsx3
from IPython.display import Audio
text = '''Hi This is Tirupathamma Agentic Ai Developer'''

audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)


voice = audio.getProperty('voices')


#audio.setProperty('voice', voice[0].id)        
audio.setProperty('voice', voice[1].id)       


audio.say(text)


audio.save_to_file(text, 'test_male_Voice.mp3')


audio.runAndWait()
```
